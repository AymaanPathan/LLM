```python
from transformers import pipeline # taking pipelines from hugging face
```

```python
getEmotion = pipeline("sentiment-analysis")  ## identifying emotions
findEmotions = getEmotion("I love using Hugging Face pipelines!")
print(findEmotions)
```

```python
generateText  = pipeline("text-generation") ## Generating Story with help of  text-generation
story = generateText("when i was 12 year old:")
print(story)
```

```python
questionAnswerPipeLine = pipeline("question-answering")
context = "google is a search engine"  ## we have to give a context on question/answer pipeline
question = "what is google"

answer = questionAnswerPipeLine(question=question, context=context)

print(answer)
```

```python
summarizePipeLine = pipeline("summarization") # summarization pipeline
wholeText = """
Hugging Face is an AI company that has gained a lot of attention in the machine learning and natural language processing (NLP) community.
They have created an open-source library called Transformers that allows users to easily load, train, and fine-tune state-of-the-art models like BERT, GPT-2, and T5.
The library has revolutionized the way researchers and developers work with NLP, making it accessible to everyone.
"""

summary = summarizePipeLine(wholeText)
print(summary)
```

```python
translator = pipeline("translation_en_to_fr") # translation pipeline

# Translate text
text = "Hello, how are you?"
result = translator(text)

print(result)
```
